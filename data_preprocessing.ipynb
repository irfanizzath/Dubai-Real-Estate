```python
# --- Binary flags ---
df['is_near_metro'] = np.where(df['nearest_metro'].notna(), 1, 0)
df['is_near_mall'] = np.where(df['nearest_mall'].notna(), 1, 0)
```

```python

# --- 2. Handle missing values ---
# filling with median per area or global median
df['instance_date'] = pd.to_datetime(df['instance_date'], errors='coerce')
df['instance_date'].fillna(df['instance_date'].median(), inplace=True)

# year and month
df['year'] = df['instance_date'].dt.year
df['month'] = df['instance_date'].dt.month

df['rooms'] = df['rooms'].fillna('Unknown')

# Property sub type NaNs
df['property_sub_type'] = df['property_sub_type'].fillna('Unknown')

def clean_room_category(x):

    x = x.strip().lower()

    # Commercial categories
    commercial_terms = ['office', 'shop', 'store', 'gym']
    if any(term in x for term in commercial_terms):
        return 'Commercial'

    # Merge similar categories
    if 'single' in x:
        return 'Studio'
    if 'studio' in x:
        return 'Studio'
    if 'penthouse' in x:
        return 'Penthouse'

    # Bedroom numbers
    for i in range(1, 11):
        if f'{i}' in x:
            return f'{i} Bedroom'

    return 'Other'


def rooms_to_numeric(x):
    mapping = {
        'Studio': 0.5,
        '1 Bedroom': 1,
        '2 Bedroom': 2,
        '3 Bedroom': 3,
        '4 Bedroom': 4,
        'Penthouse': 4,
        '5 Bedroom': 5,
        '6 Bedroom': 6,
        '7 Bedroom': 7,
        '8 Bedroom': 8,
        '9 Bedroom': 9,
        '10 Bedroom': 10,
        'Commercial': 0,
        'Other': 0, # Map 'Other' category to 0
    }
    return mapping.get(x, 0)

df['rooms'] = df['rooms'].apply(clean_room_category)
df['rooms_numeric'] = df['rooms'].apply(rooms_to_numeric)
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# ---  Merging low-count categories into 'Mixed-Use' ---
df['property_usage'] = df['property_usage'].replace({
    'Other': 'Mixed-Use',
    'Residential / Commercial': 'Mixed-Use',
    'Multi-Use': 'Mixed-Use'
})

# --- Dropping irrelevant categories ---
drop_categories = [
    'Industrial',
    'Agricultural',
    'Industrial / Commercial',
    'Storage',
    'Industrial / Commercial / Residential'
]

df = df[~df['property_usage'].isin(drop_categories)]

print(df['property_usage'].value_counts())
```

```python
df['property_sub_type'] = df.apply(
    lambda row: row['property_type'] if row['property_sub_type'] == 'Unknown' else row['property_sub_type'],
    axis=1)

subtype_mapping = {
    'Hotel Rooms': 'Hotel Apartment',
    'Stacked Townhouses': 'Villa',
    'Clinic': 'Commercial',
    'Show Rooms': 'Commercial',
    'Hotel': 'Hotel Apartment',
    'Sized Partition': 'Other',
    'Gymnasium': 'Commercial',
    'Parking': 'Commercial',
    'Unit': 'Commercial',
    'Office': 'Commercial',
    'Shop': 'Commercial'
}

# Mapping
df['property_sub_type'] = df['property_sub_type'].replace(subtype_mapping)

# Replacing Unknown with property_type_en
df['property_sub_type'] = df.apply(
    lambda row: row['property_type'] if row['property_sub_type'] == 'Unknown' else row['property_sub_type'],
    axis=1
)

print(df['property_sub_type'].value_counts())
```
